```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense, Dropout
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
df = pd.read_csv('/content/working on_Daneshga Zanjan_2019_2023csv.csv')
```

```python
df[5:]
```

```python
train_dates = pd.to_datetime(df['date'])
train_dates.tail(15)
```

```python
cols = list(df)[1:5]
cols
```

```python
df_for_training = df[cols].astype(float)
```

```python
df_for_training.head()
```

```python
scaler = StandardScaler()
scaler = scaler.fit(df_for_training)
df_for_training_scaled = scaler.transform(df_for_training)
```

```python
df_for_training_scaled.shape
```

```python
trainX = []
trainY = []
n_future = 4
n_past = 14
```

```python
for i in range(n_past, len(df_for_training_scaled) - n_future +1):
    trainX.append(df_for_training_scaled[i - n_past:i, 0:df_for_training.shape[1]])
    trainY.append(df_for_training_scaled[i + n_future - 1:i + n_future, 0])
```

```python
trainX, trainY = np.array(trainX), np.array(trainY)
```

```python
print('trainX shape == {}.'.format(trainX.shape))
print('trainY shape == {}.'.format(trainY.shape))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization

# تعریف مدل
model = Sequential()

# لایه اول LSTM
model.add(LSTM(
    units=128,
    activation='relu',
    input_shape=(trainX.shape[1], trainX.shape[2]),
    return_sequences=True,
    kernel_regularizer='l2'
))
model.add(BatchNormalization())

# لایه دوم LSTM
model.add(LSTM(
    units=64,
    activation='relu',
    return_sequences=True
))
model.add(Dropout(0.3))

# لایه Dense میانی
model.add(Dense(16, activation='relu'))  # لایه جدید

# لایه سوم LSTM (آخرین LSTM)
model.add(LSTM(
    units=32,
    activation='relu',
    return_sequences=False
))

# لایه Dropout نهایی
model.add(Dropout(0.2))

# لایه Dense خروجی
model.add(Dense(
    units=trainY.shape[1],
    activation='linear'
))

# کامپایل مدل
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# نمایش خلاصه مدل
model.summary()
```

```python
history = model.fit(trainX, trainY, epochs=50, batch_size=16, validation_split=0.1, verbose=1)
```

```python
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.show()
```
